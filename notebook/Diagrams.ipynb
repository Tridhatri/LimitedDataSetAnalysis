```python
import matplotlib.pyplot as plt

# Data
categories = [
    "Normal",
    "Anomalous (DoS attack)",
    "Anomalous (scan)",
    "Anomalous (malicious control)",
    "Anomalous (malicious operation)",
    "Anomalous (spying)",
    "Anomalous (data probing)",
    "Anomalous (wrong set up)"
]

counts = [
    48515,
    835,
    250,
    132,
    129,
    69,
    50,
    20
]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.barh(categories, counts, color='blue')
plt.xlabel('Count')
plt.title('Value Counts of Normality Column')
plt.grid(axis='x')

# Adding data labels on top of bars
for bar in bars:
    plt.text(bar.get_width() + 500, bar.get_y() + bar.get_height()/2, 
             f'{int(bar.get_width())}', 
             va='center')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
normality = [7, 0, 4, 2, 3, 5, 1, 6]
percentage = [77.63, 29.93, 8.94, 4.37, 3.95, 2.78, 1.58, 0.83]

# Create horizontal bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(normality, percentage, color='skyblue')

# Add percentages on top of the bars
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width()}%', 
             va='center', ha='left', color='black', fontsize=10)

plt.xlabel('Percentage')
plt.ylabel('Normality')
plt.title('Normality vs Percentage')
plt.show()
```
